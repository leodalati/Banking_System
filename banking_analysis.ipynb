```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("transactions.csv")
df
```

```python
df["date"] = pd.to_datetime(df["date"])
```

```python
df.info()
```

```python
df.groupby("transaction_type")["amount"].sum()
```

```python
df.sort_values("date").plot(
    x="date",
    y="balance_after",
    title="Account Balance Over Time"
)
plt.show()
```

```python
df["month"] = df["date"].dt.to_period("M")

monthly = df.groupby(["month", "transaction_type"])["amount"].sum()
monthly.unstack()
```

```python
monthly.unstack().plot(kind="bar", title="Monthly Cash Flow")
plt.show()
```
